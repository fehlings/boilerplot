# Examples

Below is a minimal example of how to use the style sheet and the effect it has.

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import boilerplot

mpl.rcParams.update(mpl.rcParamsDefault)
```

```python
x = np.linspace(0, 2*np.pi)

fig, ax = plt.subplots(1, 2, figsize=[2.5, 1.6667], dpi=300)

with mpl.style.context('boilerplot.ieeetran'):


    ax[0].plot(x, np.sin(x))
    ax[0].set_title("IEEEtran stylesheet")

ax[1].plot(x, np.sin(x))
ax[1].set_title("Default stylesheet")

plt.show()
```
