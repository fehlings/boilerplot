# Examples

Below is a minimal example of how to use the style sheet and the effect it has.

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import boilerplot

mpl.rcParams.update(mpl.rcParamsDefault)
```

```python
x = np.linspace(0, 2*np.pi)

with mpl.style.context('boilerplot.ieeetran'):

    fig_style, ax_style = plt.subplots(figsize=[3.45, 2.3], dpi=200)
    ax_style.plot(x, np.sin(x))
    ax_style.set_title("IEEEtran stylesheet")

with mpl.style.context('boilerplot.science'):

    fig_style, ax_style = plt.subplots(figsize=[3.45, 2.3], dpi=200)
    ax_style.plot(x, np.sin(x))
    ax_style.set_title("Science stylesheet")

with mpl.style.context('boilerplot.presentation'):

    fig_style, ax_style = plt.subplots(figsize=[3.45, 2.3], dpi=200)
    ax_style.plot(x, np.sin(x))
    ax_style.set_title("Presentation stylesheet")

fig_default, ax_default = plt.subplots(figsize=[3.45, 2.3], dpi=200)
ax_default.plot(x, np.sin(x))
ax_default.set_title("Default stylesheet")

plt.show()
```
