# Examples

Below is a minimal example of how to use the style sheet and the effect it has.

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

mpl.rcParams.update(matplotlib.rcParamsDefault)
```

```python
x = np.linspace(0, 2*np.pi)

with mpl.style.context('boilerplot.ieeetran'):

    fig_style, ax_style = plt.subplots()
    ax_style.plot(x, np.sin(x))
    ax_style.set_title("IEEEtran stylesheet")
    plt.show()

fig_default, ax_default = plt.subplots()
ax_default.plot(x, np.sin(x))
ax_default.set_title("Default stylesheet")
plt.show()
```
